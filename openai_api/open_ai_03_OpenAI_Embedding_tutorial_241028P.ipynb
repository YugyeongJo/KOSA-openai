```python
import openai
import numpy as np
from numpy import dot
from numpy.linalg import norm
import pandas as pd

# norm : 1 단위로 스케일링 하는 것(정규화)
```

```python
def get_embedding(text):
    response = client.embeddings.create(
    input=text,
    model="text-embedding-ada-002"
    )
    return response.data[0].embedding
```

```python
data = ['저는 배가 고파요',
        '저기 배가 지나가네요',
        '굶어서 허기가 지네요',
        '허기 워기라는 게임이 있는데 즐거워',
        '스팀에서 재밌는 거 해야지',
        '스팀에어프라이어로 연어구이 해먹을거야']
```

```python
pd.DataFrame(data, columns=['text'])
```

```python
df = pd.DataFrame(data, columns=['text'])
df
```

```python
df.apply(lambda row: get_embedding(row.text), axis=1)
```

```python
df['embedding'] = df.apply(lambda row: get_embedding(row.text), axis=1)

df
```

```python
# 여기서 잠깐
# 코사인 유사도

def cos_sim(A, B):
    return dot(A,B) / (norm(A) * norm(B))
```

```python
[1,2,3] + [10,20,30]
```

```python
a = [1,2,3]
b = [10,20,30]

np.array(a) + np.array(b)
```

```python
def return_answer_candidate(df, query):
    query_embedding = get_embedding(query)
    df['similarity'] = df.embedding.apply(lambda x: cos_sim(np.array(x),
                                               np.array(query_embedding)))
    results_co = df.sort_values('similarity',
                                ascending=False,   # 내림차순 정렬
                                ignore_index=True
                            )
    return results_co
```

```python
query = '아무 것도 안 먹었더니 꼬르륵 소리가 나네'
```

```python
return_answer_candidate(df, query)
```

```python
sim_result = return_answer_candidate(df, query)
sim_result.head(3)
```
